```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Read image
img = cv2.imread('bulan.jpg')
```

```python
#Tambahkan Gaussian noise pada gambar
mean = 0
sigma = 25
gaussian_noise = np.random.normal(mean, sigma, img.shape).astype(np.uint8)
noisy_img =cv2.add(img, gaussian_noise)
```

```python
#Terapkan gaussian blur untuk noise reduction
blurred = cv2.GaussianBlur(noisy_img, (5, 5),0)
```

```python
img_rgb= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
noisy_img_rgb = cv2.cvtColor(noisy_img, cv2.COLOR_BGR2RGB)
blurred_rgb = cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title('Real Image Bulan')
plt.axis('off')
```

```python
plt.subplot(1, 3, 2)
plt.imshow(noisy_img_rgb)
plt.title('Real Image Bulan')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(blurred_rgb)
plt.title('Real Image Bulan')
plt.axis('off')
```

```python
import cv2
from matplotlib import pyplot as plt
```

```python
#Baca Gambar grayscale
img = cv2.imread('orgil.jpg', 0)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)  # Sobel X
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)  # Sobel Y
```

```python
sobelx= cv2.convertScaleAbs(sobelx)
sobely= cv2.convertScaleAbs(sobely)
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X ( mendeteksi tepi horizontal)')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y (mendeteksi Tepi vertikal)')
plt.axis('off')
```
